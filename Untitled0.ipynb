```python
import pandas as pd
import ast
from collections import Counter

# Load file
df = pd.read_csv('imdb_data (1).csv')

# Profit & ROI
df['Profit'] = df['revenue'] - df['budget']
df['ROI'] = df['Profit'] / df['budget']

# ---------------- Q1 ----------------
print("=== Highest Profit Movie ===")

max_profit_movie = df.loc[df['Profit'].idxmax()]

print("Title:", max_profit_movie['title'])

# Director निकालना
crew = ast.literal_eval(max_profit_movie['crew'])
director = [p['name'] for p in crew if p['job'] == 'Director']
print("Director:", director)

# Actors निकालना
cast = ast.literal_eval(max_profit_movie['cast'])
actors = [p['name'] for p in cast]
print("Actors:", actors[:5])

# ---------------- Q2 ----------------
print("\n=== Highest ROI Language ===")

roi_lang = df.groupby('original_language')['ROI'].mean().sort_values(ascending=False)
print(roi_lang.head())

# ---------------- Q3 ----------------
print("\n=== Unique Genres ===")

genres_set = set()

for g in df['genres'].dropna():
    try:
        for i in ast.literal_eval(g):
            genres_set.add(i['name'])
    except:
        pass

print(genres_set)

# ---------------- Q4 ----------------
print("\n=== Top 3 Producers ===")

producer_data = []

for _, row in df.iterrows():
    try:
        crew = ast.literal_eval(row['crew'])
        for p in crew:
            if p['job'] == 'Producer':
                producer_data.append((p['name'], row['ROI']))
    except:
        pass

prod_df = pd.DataFrame(producer_data, columns=['Producer','ROI'])

print(prod_df.groupby('Producer')['ROI'].mean().sort_values(ascending=False).head(3))

# ---------------- Q5 ----------------
print("\n=== Most Frequent Actor ===")

all_actors = []

for c in df['cast'].dropna():
    try:
        cast = ast.literal_eval(c)
        for a in cast:
            all_actors.append(a['name'])
    except:
        pass

top_actor = Counter(all_actors).most_common(1)[0][0]
print(top_actor)

# ---------------- Q6 ----------------
print("\n=== Top 3 Directors ===")

directors = []

for c in df['crew'].dropna():
    try:
        crew = ast.literal_eval(c)
        for p in crew:
            if p['job'] == 'Director':
                directors.append(p['name'])
    except:
        pass

print(Counter(directors).most_common(3))
```
